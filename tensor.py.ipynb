cat/|cat| * kitty/|kitty| should have smallest value on the space of words. In other words, vector value of 'cat' and vector value of 'kitty' should be silimar on the space of words. 

```python
import tensorflow as tf

embeddings = tf.Variable(
    tf.random_uniform([5, 128], -1.0, 1.0))
norm = tf.sqrt(tf.reduce_sum(tf.square(embeddings), 1, keep_dims=True))
normalized_embeddings = embeddings / norm

print(normalized_embeddings)

with tf.Session() as session:
  tf.initialize_all_variables().run()

  print('sum of embeddings is %s' % tf.reduce_sum(embeddings[0,:]).eval(session=session))  
  print('sum of squared embeddings is %s' % tf.reduce_sum(tf.square(embeddings[0,:])).eval(session=session))    
  print('norm is %s' % norm.eval(session=session))  
    
  print('embedding ---- \n %s' % embeddings[0,:].eval(session=session))
  print('normalized embedding ---- \n  %s' % normalized_embeddings[0,:].eval(session=session))
```
